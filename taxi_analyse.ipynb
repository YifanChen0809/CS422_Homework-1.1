```python
# 绘制 fare_amount 的分布
sns.histplot(df['fare_amount'], bins=30, kde=True)
plt.title('Fare Amount Distribution')
plt.xlabel('Fare Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
#the relation between fare_amount and trip_distance(Bar Plot)
sns.scatterplot(x=df['trip_distance'], y=df['fare_amount'])
plt.title('Trip Distance vs Fare Amount')
plt.xlabel('Trip Distance')
plt.ylabel('Fare Amount')
plt.show()
```

```python
#Heatmap（分析变量之间联系的程度）
corr = df[['fare_amount', 'trip_distance', 'duration']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#payment_type vs fare_amount
sns.barplot(x='payment_type', y='fare_amount', data=df)
plt.title('Average Fare Amount by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Average Fare Amount')
plt.show()
```

```python
#prediction of fare_payment
sns.lmplot(x='trip_distance', y='fare_amount', data=df)
plt.title('Fare Amount Prediction by Trip Distance')
plt.xlabel('Trip Distance')
plt.ylabel('Fare Amount')
plt.show()
```
